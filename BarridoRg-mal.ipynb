```python
import ltspice
import numpy as np
import matplotlib.pyplot as plt
import csv
import os as os
```

\begin{align*}{V_{rms} = \sqrt{\frac{V_1^2 + V_2^2 + ... + V_n^2}{n}}}\end{align*}
\begin{align*}{
i_{rms} = \sqrt{\frac{i_1^2 + i_2^2 + ... + i_n^2}{n}}
}\end{align*}

```python
def rms(i):
    rta = 0
    for k in i:
        rta = rta + (k)**2
    return np.sqrt(rta/(len(i)))
```

```python
x = [10, 15, 18, 22, 27, 33, 47, 100]
y = []
```

```python
filepath_10 = 'Spice - Copia/CircuitRg10.raw'
l_10 = ltspice.Ltspice(filepath_10)
l_10.parse() # Data loading sequence. It may take few minutes for huge file.
t = l_10.get_time()
Vd_10 = rms(l_10.get_data("V(N001,A)"))
id_10 = rms(l_10.get_data("I(Vgau)"))
y.append(Vd_10*id_10)
plt.plot(t, l_10.get_data("I(Vgau)"))
```
